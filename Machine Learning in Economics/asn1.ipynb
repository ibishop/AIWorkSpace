#Problem 8

i) Load the data: I just downloaded it from the website and put it into working directory

```python
import pandas as pd
college=pd.read_csv('College.csv')
college.head()
```

ii) Plot scatter-plots of first ten variables

```python

from pandas.plotting import scatter_matrix
attributes = ['Private','Apps', 'Accept', 'Enroll', 'Top10perc', 'Top25perc', 'F.Undergrad','P.Undergrad','Outstate','Room.Board']
scatter_matrix(college[attributes],figsize=(6,6))
```

iii) Produce two box plots of Outstate and Private


```python
import matplotlib.pyplot as plt
%matplotlib inline

#This lines maps "Yes" and "No" to floating points, so matplotlib can actually plot them. 
college['Private'] = college['Private'].map({'Yes': 1, 'No': 0})

#subplot takes in 1-row, 2-column, as instructions, then plot number, e.g (1,2,1)

#Figure 1
plt.subplot(1, 2, 1)
plt.boxplot(college['Outstate'])

#Figure 2
plt.subplot(1, 2, 2)
plt.boxplot(college['Private'])
```

iv) Extra Column, Elite

```python

#Adding Extra column, use *1 to coerce booleans to ints
college['Elite'] =(college['Top10perc'] > 50) * 1
```

v) Histograms

```python
college.hist(bins=25,figsize=(20,15))
```

i) Load the dataset, thankfully scikit learn has this one 

```python
from sklearn import datasets
import seaborn as sns

boston = datasets.load_boston()
type(boston)
```

I need to do some work to get a useful object here. I need to seperate the bunch (which is really a dictionary) into its useful parts. There is a standard way to do this, as follows:

```python

#Standard Dataframe object comes with pandas package
X = pd.DataFrame( boston.data)

#Note, X is just data with no names at this point so we have to add names. 
X.columns = boston.feature_names
 

X.head()
```
